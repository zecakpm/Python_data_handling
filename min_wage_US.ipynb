```python
gb = df.groupby("State")
gb.get_group("Alabama").set_index("Year").head
```

```python
#setting states as the columns
act_min_wage = pd.DataFrame()

for name, group in df.groupby("State"):
    if act_min_wage.empty:
        # Setting the index as year, [[]] --> to add values as data frame
        #rename.columns --> is a dictionary --> will replace "Low.2018" to name, that in this case will be each of the examples
        act_min_wage = group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name})
    else:
        #if the tables has values, join the next interaction of the loop with previous results
        act_min_wage = act_min_wage.join(group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name}))
        
act_min_wage.head()
```
